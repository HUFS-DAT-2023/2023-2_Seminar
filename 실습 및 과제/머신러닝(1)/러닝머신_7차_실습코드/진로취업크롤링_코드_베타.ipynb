```python
# 팝업창 닫기
try :
    wd.find_element(By.XPATH, "/html/body/div[3]/div[3]/div/button[1]/span").click()
    wd.find_element(By.XPATH, '/html/body/div[3]/div[3]/div/button[1]').click()
except:
    print("error")
```

```python
# 비교과 프로그램 리스트로 이동 - 진로.취업
wd.find_element(By.XPATH,'//*[@id="mainWidgetUseLstWrap"]/li[4]/div/a').click()
workshop_link = wd.current_url # 특강및워크숍카테고리 url 저장
workshop_driver = wd # 특강및워크숍 카테고리 드라이버 저장
time.sleep(1)
```

```python
# 전체 페이지가 70페이지 정도, 각 페이지를 한 장 씩 넘기며 해당 페이지에 있는 프로그램에 접속
now_page_num = 0
df = pd.DataFrame()
for total_page_cnt in range(7):
    for page_num in range(3,13):
        now_page_num += 1
        print(now_page_num)
        wd.find_element(By.XPATH,'/html/body/div[2]/div[2]/section/div/div[2]').find_elements(By.TAG_NAME, 'li')[page_num].click()
        time.sleep(1)
        # 각 프로그램 크롤링
        for i in range(10):
            i = i*3+1# 프로그램 링크 인덱스 이상해서 그거 맞춰서 인덱싱
            titles = wd.find_elements(By.CLASS_NAME, 'detailBtn')
            title_text = titles[i].text
            try :
                titles[i].click()
            except:
                print("now page num = " , now_page_num ," ", i ,'번째 프로그램 titles click error')
            # title_text = driver.find_element(By.CLASS_NAME, "tab_top_wrap").text
            contents = wd.find_element(By.TAG_NAME, "tbody")
            content_text_dict = {key.text : [value.text] for key, value in zip(contents.find_elements(By.TAG_NAME, "th"),contents.find_elements(By.TAG_NAME, "td"))}
            # 프로그램 데이터 추출
            df_row = pd.concat([pd.DataFrame({'프로그램명':[title_text]}),pd.DataFrame(content_text_dict)], axis = 1)
            # print(df_row[['프로그램 분류','선정방법']])
            # df에 행 추가
            df = pd.concat([df, df_row])
            time.sleep(1)
            # 뒤로가기
            wd.back()
            time.sleep(1)
```
