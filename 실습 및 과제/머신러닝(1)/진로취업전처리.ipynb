```python
employee=employee.drop(['Unnamed: 0','첨부파일'], axis=1, errors='ignore')
```

```python
employee[['시작날짜', '마감날짜']] = employee['신청기간'].str.split(' ~ ', expand=True)
employee['시작날짜'] = pd.to_datetime(employee['시작날짜'])
employee['마감날짜'] = pd.to_datetime(employee['마감날짜'])
employee2=employee.drop(['신청기간'], axis=1, errors='ignore')
```

```python
employee2['정원'] = employee2['정원/대기자'].str.extract(r'정원 (\d+)명')
employee2['대기자'] = employee2['정원/대기자'].str.extract(r'대기자 (\d+)명')

employee3=employee2.drop(['정원/대기자'], axis=1, errors='ignore')
```

```python
employee3['학부'] = employee3['신청조건'].str.extract(r'학부 : ([^\n]+)')
employee3['학적'] = employee3['신청조건'].str.extract(r'학적 : ([^\n]+)')
employee3['학년'] = employee3['신청조건'].str.extract(r'학년 : ([^\n]+)')

employee4=employee3.drop(['신청조건'],axis=1)
```

```python
employee4['수료 인증서'] = employee4['수료 인증서'].replace({'O': 1, 'X': 0})
employee4['수료 인증서'] = employee4['수료 인증서'].fillna(0)
employee4['수료 인증서'] = employee4['수료 인증서'].astype(int)
```

```python
import re

def split_text_by_percentage_and_numbers(input_series):
    new_df = pd.DataFrame()

    for i, text in input_series.items():
        if pd.isna(text) or isinstance(text, (int, float)):
            text = '0'  # NaN 값 처리
        else:
            text = str(text)

        parts = text.split('%')

        skills = {}
        for part in parts:
            part = part.strip()
            parts_with_numbers = re.split(r'(\d+)', part)
            parts_with_numbers = [part for part in parts_with_numbers if part.strip()]
            for j in range(0, len(parts_with_numbers), 2):
                if j + 1 < len(parts_with_numbers):  # 인덱스 에러 방지
                    skills[parts_with_numbers[j]] = int(parts_with_numbers[j+1])

        temp_df = pd.DataFrame(skills, index=[i])
        new_df = pd.concat([new_df, temp_df], axis=0)

    return new_df

# workshop4 DataFrame에서 '핵심역량 비중도' 열 사용
extracted_data = split_text_by_percentage_and_numbers(employee4['핵심역량 비중도'])

# NaN 값을 0으로 대체하고 모든 데이터 값을 정수로 변환
extracted_data = extracted_data.fillna(0).astype(int)

# 새로운 DataFrame에 합쳐서 저장
employee4 = pd.concat([employee4, extracted_data], axis=1)

employee5=employee4.drop(['핵심역량 비중도'],axis=1)

employee5
```

```python
def process_text(text):
    if isinstance(text, str):
        # '-'를 기준으로 앞 부분 제거
        processed_text = text.split('-', 1)[-1].strip()

        # 특수문자 제거
        processed_text = re.sub(r'[^\w\s]', '', processed_text)

        return processed_text
    return text

# '특강 및 워크숍 - 교육.특강' 열의 데이터 처리
employee5['프로그램 분류'] = employee5['프로그램 분류'].apply(process_text)
```

```python
def get_keyword(sentence):
    if pd.isnull(sentence):
        return 0
    doc = sentence
    kw_model = KeyBERT()
    keywords = kw_model.extract_keywords(doc,keyphrase_ngram_range=(0,1),use_maxsum = False,top_n = 10)
    result = [x[0] for x in keywords]
    return result
```

```python
import re

def remove_special_characters(text):
    if isinstance(text, str):
        return re.sub(r'[^\w\s]', '', text)
    return text

# '프로젝트명' 열에 특수문자 제거 적용
df3['프로그램명'] = df3['프로그램명'].apply(remove_special_characters)
df3['프로그램 주요내용'] = df3['프로그램 주요내용'].apply(remove_special_characters)
```

```python
def remove_brackets(text):
    if isinstance(text, str):
        return text.replace('[', '').replace(']', '').replace("'", '')
    return text

df3['프로그램 목표_키워드'] = df3['프로그램 목표_키워드'].apply(remove_brackets)
df3['프로그램 주요내용_키워드'] = df3['프로그램 주요내용_키워드'].apply(remove_brackets)
```
