### 1. 데이터 불러오기

```python
import pandas as pd
data = pd.read_csv('data.csv')
data
```

### 2. 데이터 정제

#### 2-1. 결측치 제거하기 & 삭제된 행 확인하기

```python
true_data = data.dropna()
true_data
```

```python
deleted_data = data[~data.index.isin(true_data.index)]
deleted_data
```

6개의 행이 삭제됨

#### 2-2. '-' 제거하기 & 삭제된 행 확인하기

```python
pitcher = true_data[~true_data.apply(lambda row: row.astype(str).str.contains('-')).any(axis=1)]
pitcher
```

```python
deleted_pitcher = true_data[~true_data.index.isin(pitcher.index)]
deleted_pitcher
```

53개의 행이 삭제됨

```python
rows_with_null = pitcher[pitcher.isnull().any(axis=1)]
rows_with_null
```

#### 2-3. 신장 변수의 cm, 체중 변수의 kg 제거하기

```python
pitcher['신장'] = pitcher['신장'].str.replace('cm', '').astype(float)
pitcher['체중'] = pitcher['체중'].str.replace('kg', '').astype(float)

pitcher[['신장', '체중']].head()
```

### 6. 이상치, 결측치 제거하기

#### 6.1 경기수 < 10인 선수 제거

```python
pitcher = pitcher[pitcher['G'] >= 5]
pitcher
```

#### 6.2 각 변수에 대한 상자그림 그리기

##### 6.2.1 신장, 체중, 방어율, 승, 패

```python
pitcher['ERA'] = pitcher['ERA'].astype('float64') #ERA 값들이 문자형이므로 실수형으로 변경
pitcher['ERA'].dtypes
```

```python
import matplotlib.pyplot as plt

columns_to_plot = ['height', 'weight', 'ERA', 'W', 'L'] 

# 이상치 삭제 전 상자그림
plt.figure(figsize=(15, 10))

for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(2, len(columns_to_plot), i)
    pitcher[[column]].boxplot()
    plt.title(f'Boxplot of {column}')
    plt.ylabel('Values')
```

```python
# 각 열에 대한 이상치 식별과 삭제
columns_to_clean = ['height', 'weight', 'ERA', 'W', 'L']

for column in columns_to_clean:
    Q1 = pitcher[column].quantile(0.25)
    Q3 = pitcher[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # 이상치 삭제 후 새로운 데이터프레임 생성
    pitcher = pitcher[(pitcher[column] >= lower_bound) & (pitcher[column] <= upper_bound)]
```

```python
# 이상치 삭제 후 상자그림

plt.figure(figsize=(15, 10))

for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(2, len(columns_to_plot), i + len(columns_to_plot))
    Q1 = pitcher[column].quantile(0.25)
    Q3 = pitcher[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    pitcher_no_outliers = pitcher[(pitcher[column] >= lower_bound) & (pitcher[column] <= upper_bound)]
    pitcher_no_outliers[[column]].boxplot()
    plt.title(f'Boxplot of {column}')
    plt.ylabel('Values')

plt.tight_layout()
plt.show()
```
